```python
import os
import sys
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

sys.path.append('../')
from pos_embedding import *
```

```python
pos_arr = np.arange(200)
d_model = 6
emd_out = get_1d_sincos_pos_embed_from_grid(d_model, pos_arr, omega_denom=100)

fig, axes = plt.subplots(d_model, 1, figsize=(12, 2 * d_model))

for i, ax in enumerate(axes):
    ax.plot(pos_arr, emd_out[:, i])
    ax.set_ylim(-1, 1)
    ax.set_xlim(0, len(pos_arr))
    ax.plot([0, len(pos_arr)], [0, 0], "k:")
```

```python
embed_dim = 8
grid_size = 14
emb = get_2d_sincos_pos_embed(embed_dim, grid_size, cls_token=False, omega_denom=25)


fig, axes = plt.subplots(embed_dim // 4, 4,  figsize=(18, 1 * embed_dim))

for i in range(embed_dim // 4):
    for j in range(4):
        axes[i, j].imshow(
            emb.reshape([grid_size, grid_size, embed_dim])[:, :, i * 4 + j],
            vmin=-1.0,
            vmax=1.0,
            cmap="RdYlBu",
        )
        axes[i, j].set_xticks([])
        axes[i, j].set_yticks([])
```
